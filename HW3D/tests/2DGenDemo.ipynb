```python
import testgen
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 200
```

```python
DEMO_ROWS = 5
DEMO_COLS = 3
```

```python
def to_poly(triangle):
    x = []
    y = []
    z = []
    for p in triangle:
        x.append(p[0])
        y.append(p[1])
        z.append(p[2])
    return x, y, z
```

```python
def draw_triangle(subplot, triangle, color, zorder):
    triangle = to_poly(triangle)
    poly = Polygon(np.column_stack([triangle[0], triangle[1]]), color=color, zorder=zorder)
    subplot.add_patch(poly)
    subplot.plot(triangle[0] + [triangle[0][0]], triangle[1] + [triangle[1][0]], color="black", zorder=3)
```

```python
def draw_example(subplot, gen_foo):
    triangle = testgen.gen(gen_foo)
    base = testgen.create_base()
    draw_triangle(subplot, base, "blue", 1)
    draw_triangle(subplot, triangle, "orange", 2)
    
    subplot.set_xlim(testgen.xlim())
    subplot.set_ylim(testgen.ylim())
    subplot.grid()
```

```python
def draw_demo(gen_foo, title):
    fig, axs = plt.subplots(DEMO_ROWS, DEMO_COLS, figsize=(15, 15))
    
    for i in range(DEMO_ROWS):
        for k in range(DEMO_COLS):
            draw_example(axs[i][k], gen_foo)
    plt.suptitle(title, fontsize=30)
    plt.show()
```

```python
draw_demo(testgen.A1, "A1 intersection case")
```
